```python
import sys  
sys.path.insert(0, '../src')
from dnns2 import *
```

```python
""" Completed:
    '1M-LdSt2.csv', 'compute_fp_1-LdSt2.csv', 'compute_fp_4-LdSt2.csv', 'compute_fp_9-LdSt2.csv', \
          'compute_fp_10-LdSt2.csv',
""" 
files = [ 'compute_int_0-LdSt2.csv', 'compute_int_4-LdSt2.csv', \
          'compute_int_7-LdSt2.csv', 'compute_int_11-LdSt2.csv', 'compute_int_15-LdSt2.csv',\
          'srv_10-LdSt2.csv']

results = pd.DataFrame(columns = ['Error'])

w=90
r=0.00075
e=20

for file in files:
    print(file) # So we see some progress while its running
    # Prepare the data for the N.N.
    df = prepDataFrame("../data/" + file)
    num_training_examples = int(0.8*len(df))
    # Build and train model
    X_tr, X_te, y_tr, y_te = test_train_split(df, num_training_examples)
    model = build_model(do1=w, do2=w, do3=w, do4=w, do5=w, l_rate=r)
    model.fit(X_tr, y_tr, epochs=e)
    # Make predictions, compute error and record the error
    predictions = model.predict(X_te)
    rmse = get_rmse(predictions, y_te, y_te.shape[0])
    results.loc[file] = [rmse]
```
